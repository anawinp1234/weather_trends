```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = '/Users/anawinp/Projects/weather_trends/data/weather_dataset_19-01-2020.csv'
df = pd.read_csv(file, index_col = 0)
df.head(10)
```

```python
df.info()
df.describe()
df.isna().sum()
df = df.dropna()
```

```python
df = df.rename(columns={'temparature': 'temperature'})
df.info()
# df['temperature'] = df['temperature'].str.strip('°C')
# df['pressure'] = df['pressure'].str.strip('mbar')
# df['pressure'] = df['pressure'].str.strip()
df['pressure'] = df['pressure'].astype(int)
df['dev_point'] = df['dev_point'].str.replace('°C', '', regex=False)
# df['feels_like'] = df['feels_like'].str.replace('°C', '', regex=False)
df = df[df['pressure'] != 'N/A']
df['visibility'] = df['visibility'].str.replace('N/A', '0', regex=False)
df['dev_point'] = df['dev_point'].str.replace('N/A', '0', regex=False)
df['visibility'] = df['visibility'].str.replace('km', '', regex=False)
df['visibility'] = df['visibility'].str.strip('\xa0')
df['dev_point'] = df['dev_point'].str.strip('\xa0')
# df['feels_like'] = df['feels_like'].str.strip('\xa0')
df['feels_like'] = df['feels_like'].astype(int)
df['dev_point'] = df['dev_point'].astype(int)
df['visibility'] = df['visibility'].astype(int)
df.info()
df
```

```python

g = sns.relplot(x='current_time', y='temperature', data=df, kind='line')
g.set(xlabel='Time', ylabel='Temperature')
g.fig.suptitle('Temperature over Time', y=                   1.03 )
plt.xticks(rotation=68)
plt.show()
```
